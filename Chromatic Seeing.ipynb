```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font', size=20)
mpl.rc('figure', figsize=(12,8))

import numpy as np
import cPickle
```

Kolmogorov turbulence in the atmosphere induces a $\lambda^{-1/5}$ dependence of the PSF FWHM on wavelength.  We want to see this in SDSS.  We don't want to get confused by differential chromatic refraction though, so we limit our observations to near zenith where DCR should be negligible.  The following CasJobs command selects the 50000 suitable stars closest to zenith in the DR12 database:

`
SELECT TOP 50000
  objid, run, rerun, probPSF, ra, dec,
  cModelMag_u, cModelMag_g, cModelMag_r, cModelMag_i, cModelMag_z,
  cModelMagErr_u, cModelMagErr_g, cModelMagErr_r, cModelMagErr_i, cModelMagErr_z,
  mRrCc_u, mRrCc_g, mRrCc_r, mRrCc_i, mRrCc_z,
  mRrCcErr_u, mRrCcErr_g, mRrCcErr_r, mRrCcErr_i, mRrCcErr_z,
  mE1_u, mE1_g, mE1_r, mE1_i, mE1_z,
  mE2_u, mE2_g, mE2_r, mE2_i, mE2_z,
  mE1E1Err_u, mE1E1Err_g, mE1E1Err_r, mE1E1Err_i, mE1E1Err_z,
  mE1E2Err_u, mE1E2Err_g, mE1E2Err_r, mE1E2Err_i, mE1E2Err_z,
  mE2E2Err_u, mE2E2Err_g, mE2E2Err_r, mE2E2Err_i, mE2E2Err_z,
  mCr4_u, mCr4_g, mCr4_r, mCr4_i, mCr4_z,
  mRrCcPSF_u, mRrCcPSF_g, mRrCcPSF_r, mRrCcPSF_i, mRrCcPSF_z,
  mE1PSF_u, mE1PSF_g, mE1PSF_r, mE1PSF_i, mE1PSF_z,
  mE2PSF_u, mE2PSF_g, mE2PSF_r, mE2PSF_i, mE2PSF_z,
  mCr4PSF_u, mCr4PSF_g, mCr4PSF_r, mCr4PSF_i, mCr4PSF_z,
  airmass_u, airmass_g, airmass_r, airmass_i, airmass_z,
  phioffset_u, phioffset_g, phioffset_r, phioffset_i, phioffset_z,
  offsetRa_u, offsetRa_g, offsetRa_r, offsetRa_i, offsetRa_z,
  offsetDec_u, offsetDec_g, offsetDec_r, offsetDec_i, offsetDec_z,
  TAI_u, TAI_g, TAI_r, TAI_i, TAI_z
FROM Star
INTO StarShapeZenith2
WHERE 
  cModelMagErr_g < 0.01
  AND airmass_g BETWEEN 1.00000 and 1.004
  AND ((flags_r & 0x10000000) != 0)
  AND ((flags_r & 0x8100000c00a4) = 0)
  AND (((flags_r & 0x400000000000) = 0) OR (psfmagerr_r <= 0.2))
  AND (((flags_r & 0x100000000000) = 0) OR (flags_r & 0x1000) = 0)
  AND ((calibStatus_u & 1) != 0)
  AND ((calibStatus_g & 1) != 0)
  AND ((calibStatus_r & 1) != 0)
  AND ((calibStatus_i & 1) != 0)
  AND ((calibStatus_z & 1) != 0)
ORDER BY airmass_g
`

The "star_data.pkl" file is created using code from the github.com/jmeyers314/chroma.git repo.  It is used to create predictions for DCR and chromatic seeing effects given a library of stellar spectra, which in this case is LSST CatSim.

Color-color sanity check
========================

The color-color diagram should be a good sanity check of the CatSim spectra compared to data.  There's a pretty good match in the plot below, with a smallish offset in the M-star column g-r color.  The SDSS data also appear to be smeared diagonally, which for now I'm blaming on dust.  Shouldn't matter much for the chromatic tests further below, since both the color and chromatic effect in question will be affected by dust.

```python
# Color-color diagram
plt.figure()
plt.scatter(data['cModelMag_g'] - data['cModelMag_r'], 
            data['cModelMag_r'] - data['cModelMag_i'],
            alpha=0.02)
plt.scatter(sim['magCalc']['SDSS_g'] - sim['magCalc']['SDSS_r'],
            sim['magCalc']['SDSS_r'] - sim['magCalc']['SDSS_i'],
            color='cyan', s=1)
plt.xlim(-0.8,2.5)
plt.ylim(-0.8,2)
plt.xlabel("g-r")
plt.ylabel("r-i")
plt.show()
```

Since chromatic seeing is a multiplicative effect on the PSF size, but the absolute size may vary quite a bit from night to night, it's easier to look at the relative PSF size. 

To measure this effect in SDSS data, we compare the individual measurements of stellar PSF second moment size (Ixx + Iyy) to the PSF model second moment size at the positions of said stars.  Since the model is constructed from stars of many colors, we expect that bluer stars will have a positive residual and redder stars will have a negative residual.  We again compare to CatSim predictions, but note that there is an unknown vertical offset since the distribution of stars used to form the SDSS PSF model may not exactly match the distribution of stars in CatSim.

A slope is significantly detected, but appears significantly steeper than the CatSim model predictions.  

One potential source of the discrepancy is in the definition of second moments, which are apodized in the case of the SDSS data, but unapodized for the model.  Since the moments are normalized to the also-apodized-flux, it's tricky to determine if this will lead to larger- or smaller-than-expected PSF sizes.

```python
# Look at PSF size residual
xsim = sim['magCalc']['SDSS_g'] - sim['magCalc']['SDSS_r']
ysim = (sim['S_m02']['SDSS_g'] - sim['S_m02']['SDSS_g'].mean())/sim['S_m02']['SDSS_g'].mean()

plt.figure(figsize=(18,20))
for i, za in enumerate(np.arange(1.0, 4.1, 1.0)):
    w = (data['za_g'] < za) & (data['za_g'] > za-1.0)
    xdata = (data['cModelMag_g'] - data['cModelMag_r'])[w]
    ydata = ((data['mRrCc_g'] - data['mRrCcPSF_g'])/data['mRrCcPSF_g'])[w]

    # binned results in different percentiles of color
    pcs = [np.percentile(data['cModelMag_g'] - data['cModelMag_r'], x) for x in xrange(0, 101, 10)]
    xmean = []
    ymean = []
    xerr = []
    yerr = []
    for j in range(1, len(pcs)):
        ww = (xdata<pcs[j]) & (xdata>pcs[j-1])
        xmean.append(xdata[ww].mean())
        xerr.append((xmean[j-1] - pcs[j-1], pcs[j] - xmean[j-1]))
        ymean.append(ydata[ww].mean())
        yerr.append(ydata[ww].std()/np.sqrt(ww.sum()))
        #yerr.append(ydata[ww].std())

    plt.subplot(4, 2, i+1)
    plt.ylim(-0.1, 0.1)
    plt.xlim(0.0, 1.6)
    plt.axhline(0, color='k', alpha=0.3)
    
    plt.scatter(xdata, ydata, alpha=0.02*(5-i))
    plt.scatter(xsim, ysim, color='cyan', s=1)
    plt.errorbar(xmean, ymean, ls=' ', lw=2, xerr=zip(*xerr), yerr=yerr, c='r')

    plt.xlabel(r"g-r")
    plt.ylabel(r"$\Delta r^2 / r^2$")
    plt.title(r"{}$^\circ$ < za < {}$^\circ$".format(za-1.0, za))
plt.tight_layout()
plt.show()
```
